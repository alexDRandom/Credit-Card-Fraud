```python
print('No Frauds', round(df['Class'].value_counts()[0]/len(df) * 100,2), '% of the dataset')
print('Frauds', round(df['Class'].value_counts()[1]/len(df) * 100,2), '% of the dataset')
```

```python
import matplotlib.pyplot as plt

# Plot transaction count over time
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Amount'])
plt.xlabel('Time')
plt.ylabel('Amount')
plt.title('Transaction Amount Over Time')
plt.show()
```

```python
import seaborn as sns
```

```python
# Histogram of a numeric feature
plt.hist(df['V1'], bins=20)
plt.xlabel('V1')
plt.ylabel('Frequency')
plt.title('Histogram of V1')
plt.show()

# Box plot for Amount by Class
sns.boxplot(x='Class', y='Amount', data=df)
plt.xlabel('Class (0: Non-fraud, 1: Fraud)')
plt.ylabel('Amount')
plt.title('Box Plot of Amount by Class')
plt.show()

# Scatter plot of V1 vs. V2
plt.scatter(df['V1'], df['V2'], c=df['Class'], cmap='coolwarm')
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('Scatter Plot of V1 vs. V2 (Color-coded by Class)')
plt.show()
```

```python
# Bar plot of Class distribution
class_counts = df['Class'].value_counts()
class_counts.plot(kind='bar')
plt.xticks([0, 1], ['Non-fraud', 'Fraud'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Transactions by Class')
plt.show()
```
